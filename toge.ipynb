**Expected**

name, mv or audio, view, like

```python
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from chromedriver_py import binary_path # this will get you the path variable
from bs4 import BeautifulSoup
import xlsxwriter
import lxml
import pandas as pd
```

```python
#Read file
toge_df = pd.read_excel("C:\\Users\\RRTT\\Desktop\\Project\\First Project\\togenashi_togeari_links.xlsx", index_col= False)
print(toge_df.head())
```

```python
#Check on mv data
toge_mv = toge_df[['en_name','mv_link']]
toge_mv = toge_mv.dropna() #Remove null links

urls_mv = list(toge_mv['mv_link']) #Create a list of links
```

```python
service = Service(binary_path)
driver = webdriver.Chrome(service=service)

t_mv = []
v_mv = []
l_mv =[]
for url in urls_mv:
    driver.get(url)

    wait = WebDriverWait(driver, 10)

    # --- Title ---
    title = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "h1.ytd-watch-metadata"))).text
    t_mv.append(title)

    # --- Views ---
    views = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, 'span.style-scope.yt-formatted-string.bold'))).text
    v_mv.append(views)

    # --- Likes ---
    likes_button = wait.until(EC.presence_of_element_located((
        By.CSS_SELECTOR,
        "button[title='I like this']"
    )))

    #Full number ("like this video along with 28,235 other people")
    likes_label = likes_button.get_attribute("aria-label")

    l_mv.append(likes_label)
```

```python
#Check the result
print(t_mv[0])
print(v_mv[0])
print(l_mv[0])
```

```python
#Combine into DataFrame
#Change view into number format
v_mv_new = []
for vw in v_mv:
    if 'M' in vw:
        new_vw = float(vw[:vw.find("M")])*1000000
    
    elif 'K' in vw:
        new_vw = float(vw[:vw.find("K")])*1000

    v_mv_new.append(new_vw)

#Extract like into number format
l_mv_new = []
for li in l_mv:
    new_li = li[li.find('with ')+5:li.find(' other')]
    new_li = new_li.replace(',','')
    l_mv_new.append(int(new_li))

#Combine all data into DataFrame
mv_df = pd.DataFrame({'Title':t_mv,
                      'url':urls_mv,
                      'type':'MV',
                      'view':v_mv_new,
                      'like':l_mv_new})
#Check the result
mv_df.head()
```
